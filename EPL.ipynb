```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
epl = pd.read_csv('england-epl-teams-2018-to-2019-stats.csv')
epl
```

```python
epl.dtypes
```

```python
#1- Average possession percentage for Crystal Palace in the 2018/2019 season?
print('1- Average possession percentage for Crystal Palace in the 2018/2019 season?')

avg = epl['average_possession'].mean()
print('Answer:')
print('The average possession percentage for Crystal Palace in the 2018/2019 season was :', avg)

```

```python
print('Bar graph for question 1')

# Plotting a bar chart
plt.figure(figsize=(10, 6))
plt.bar(epl['team_name'], epl['average_possession'], color='skyblue')

# Highlight Crystal Palace's average possession percentage
plt.bar( 'Crystal Palace' , avg , color='orange', label=f'Crystal Palace Avg: {avg}%')

plt.xlabel('Team')
plt.ylabel('Average Possession Percentage')
plt.title('Average Possession Percentage of Premier League Teams (2018/2019)')
plt.xticks(rotation=90)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#3- What was Leicester City's league position at the end of the season?
print("3- What was Leicester City's league position at the end of the season?")

Leicester_City = epl[epl['team_name']=='Leicester City FC']
league_position_leicester_city = Leicester_City['league_position'].values[0]


print('Answer:')
print(f'At the end of the season, Leicester City was in the {league_position_leicester_city}th position')

```

```python
#4- Which team had the highest number of clean sheets in away matches?
print("4- Which team had the highest number of clean sheets in away matches?")

epl_cl = epl.loc[epl['clean_sheets_away'].idxmax()]
team_with_max_cl = epl_cl['team_name']

print('Answer:')
print('The team with the highest number of clean sheets in away matches is: ',team_with_max_cl )
```

```python
#5- Total number of corners per match for Everton?
print("5- Total number of corners per match for Everton?")

 
Everton = epl[epl['team_name'] == 'Everton FC']
everton_corners = Everton['corners_total'].sum()
everton_matches  = Everton['matches_played'].sum()

if everton_matches > 0 :
    corners_per_matches = everton_corners / everton_matches
else :
    corners_per_matches = 0


print('Answer:')
print("The total nb of corners per match for everton is:" , corners_per_matches,"corners")
```

```python
#6- How many matches did Manchester City lose in the 2018/2019 season?
print("6- How many matches did Manchester City lose in the 2018/2019 season?")

manchester_city  = epl[epl['team_name'] == 'Manchester City FC']
manchester_city_losses = manchester_city['losses'].values[0]


print('Answer:')
print('Manchester City loses in the 2018/2019 season' , manchester_city_losses , 'matches' )
```

```python
#7- What was the goal difference for Burnley in home matches?
print("7- What was the goal difference for Burnley in home matches?")

Burnley = epl[epl['team_name'] == 'Burnley FC']
Burnley_goal_difference = Burnley['goal_difference_home'].values[0]

print('Answer:')
print("The goal difference for Burnley in home matches is:" , Burnley_goal_difference , 'goals')

```

```python
#8- Which team had the highest number of points per game at half-time?
print("8- Which team had the highest number of points per game at half-time?")

#highest_nb = epl['points_per_game_half_time'].max()
highest_nb = epl.loc[epl['points_per_game_half_time'].idxmax()]
team_highest_nb = highest_nb['team_name']

print('Answer:')
print("The team that had the highest number of points per game at half-time is :", team_highest_nb)
```

```python
#9- How many goals did Arsenal concede in the first 10 minutes of matches?
print("9- How many goals did Arsenal concede in the first 10 minutes of matches?")

arsenal = epl[epl['team_name'] == 'Arsenal FC']
arsenal_goals_first10min = arsenal['goals_scored_min_0_to_10'].values[0]

print('Answer:')
print("Arsenal has concede to score " , arsenal_goals_first10min , "goals in the first 10 mins")
```

```python
#10- Leicester City's average goals scored per match in the 2018/2019 season?
print("10- Leicester City's average goals scored per match in the 2018/2019 season?")

Leicester_city = epl[epl['team_name'] == 'Leicester City FC']
avg = Leicester_city['goals_scored'] / Leicester_city['matches_played']
Leicester_city_avg_goals = avg.values[0] 

print('Answer:')
print("Leicester City's average goals scored per match in the 2018/2019 season is:", Leicester_city_avg_goals, 'goals')
```
